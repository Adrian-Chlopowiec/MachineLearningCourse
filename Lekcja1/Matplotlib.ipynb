```python
import matplotlib.pyplot as plt
import numpy as np
```

# Rysowanie wykresu punktowego

```python
x = np.arange(1, 10, 1)
y = np.log(x)
plt.scatter(x, y)
plt.show()
```

# Rysowanie wykresu

```python
x = np.arange(1, 10, 1)
y = np.log(x)
plt.plot(x, y)
plt.show()
```

```python
# im więcej będzie punktów, tym dokładniejszy będzie wykres
x = np.arange(1, 10, 0.01)
y = np.log(x)
plt.plot(x, y)
plt.show()
```

# Rysowanie wykresu słupkowego

```python
labels = ['kotki', 'pieski', 'rybki']
heights = [92, 120, 7]
plt.bar(labels, heights)
plt.show()
```

# Dostosowywanie wykresu

```python
# do naszego wykresu możemy dodać tytuł
labels = ['kotki', 'pieski', 'rybki']
heights = [92, 120, 7]
plt.bar(labels, heights)
plt.title('Wyniki ankiety: "Jakie zwierze masz w domu?"')
plt.show()
```

```python
# możemy dodać kolor...
labels = ['kotki', 'pieski', 'rybki']
heights = [92, 120, 7]
plt.bar(labels, heights, color='red')
plt.title('Wyniki ankiety: "Jakie zwierze masz w domu?"')
plt.show()
```

```python
# ... a nawet różne kolory dla kolejnych słupków
labels = ['kotki', 'pieski', 'rybki']
heights = [92, 120, 7]
plt.bar(labels, heights, color=['red', 'purple', 'green'])
plt.title('Wyniki ankiety: "Jakie zwierze masz w domu?"')
plt.show()
```

```python
# możemy dodać siatkę
x = np.arange(1, 10, 0.01)
y = np.log(x)
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
# możemy ustalić granice wartości na osiach x oraz y
x = np.arange(1, 10, 0.01)
y = np.log(x)
plt.plot(x, y)
plt.grid()
plt.xlim(1, 10)
plt.ylim(-1, 5)
plt.show()
```

```python
# możemy nadać tytuły osiom
x = np.arange(1, 10, 0.01)
y = np.log(x)
plt.plot(x, y)
plt.grid()
plt.title('Log(x)')
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.show()
```

```python
# możemy użyć markerów, aby zaznaczyć policzone przez nas punkty (wykres słupkowy i plot jednocześnie)
x = np.arange(1, 10, 1)
y = np.log(x)
plt.plot(x, y, marker='o', markersize=10)
plt.show()
```

# Rysowanie wielu funkcji na jednym wykresie

```python
x = np.arange(-5, 5.001, 0.1)
y_sin = np.sin(x)
y_log = np.log(x)
plt.plot(x, y_sin, color='red')
plt.plot(x, y_log, color='green')
plt.grid()
plt.show()
```

```python
# możemy dodać do takiego obrazka etykiety
x = np.arange(-5, 5.001, 0.1)
y_sin = np.sin(x)
y_log = np.log(x)
plt.plot(x, y_sin, color='red', label='sin(x)')
plt.plot(x, y_log, color='green', label='log(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
# wypadałoby podkreślić osie x oraz y
x = np.arange(-5, 5.001, 0.1)
y_sin = np.sin(x)
y_log = np.log(x)
plt.plot(x, y_sin, color='red', label='sin(x)')
plt.plot(x, y_log, color='green', label='log(x)')
plt.grid()
plt.axhline(color='black', linewidth=1)
plt.axvline(color='black', linewidth=1)
plt.legend()
plt.show()
```
