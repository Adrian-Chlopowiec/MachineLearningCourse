# Zadanie 2

```python
# tworzę funkcję
def f(x):
    return np.exp(np.sin(x)) + np.sqrt(x)
```

```python
# tworzę funkcję pochodną
def df(x):
    return (1 / (2 * np.sqrt(x))) + np.exp(np.sin(x)) * np.cos(x)
```

```python
# rysuję wykres, ma być od 0 do 15
# wybieram wartości co 0.01, więc górna granica musi być większa od 15.01 i mniejsza niż 15.02
x = np.arange(0, 15.011, 0.1)
y_f = f(x)
y_df = df(x)
# rysuję wykres
plt.plot(x, y_f, color='orange', label='f(x)')
plt.plot(x, y_df, color='green', label='df(x)')
plt.grid()
plt.legend()
plt.axhline(color='black', linewidth=1)
plt.axvline(color='black', linewidth=1)
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.show()
# oczywiście można to wykonać na kilka sposóbów
```

# Zadanie 4

```python
# normalnie używalibyśmy pliku csv, ale jeszcze ich nie znamy, robimy więc ręcznie
brands = ['Harnaś', 'Żubr', 'Perła', 'Tatra']
sales = np.array([[160, 100, 180, 41],
                  [221, 176, 120, 240],
                  [15, 0, 162, 12],
                  [52, 71, 68, 62]])
# zdefiniujmy też używane kolory
colors = ['Red', 'Orange', 'Green', 'Blue']
```

```python
# stwórzmy funkcję, aby nie powielać kodu
def create_shop_plot(title, data, labels=brands, colors=colors):
    plt.title(title)
    plt.bar(labels,
            data,
            color=colors)
    plt.grid()
    plt.show()
```

```python
# pierwszy wiersz to pierwszy sklep
create_shop_plot('Dane sprzedażowe w sklepie Kijanka', sales[0])
```

```python
create_shop_plot('Dane sprzedażowe w sklepie Wyspa', sales[1])
```

```python
create_shop_plot('Dane sprzedażowe w sklepie Zakupoświat', sales[2])
```

```python
create_shop_plot('Dane sprzedażowe w sklepie Żuczek', sales[3])
```

```python
# dane dotyczące wszystkich sklepów
# użyjemy funkcji z biblioteki numpy: sum
# najpierw sumujemy po wszystkich kolumnach:
sales_brand = np.sum(sales, axis=0)
create_shop_plot('Dane sprzedażowe wszystkich marek', sales_brand)
```

```python
# teraz po wierszach
sales_shop = np.sum(sales, axis=1)
# tworzymy nowe labelki
shop_labels = ['Kijanka', 'Wyspa', 'Zakupoświat', 'Żuczek']
create_shop_plot('Dane sprzedażowe wszystkich sklepów', sales_shop, labels=shop_labels)
```
